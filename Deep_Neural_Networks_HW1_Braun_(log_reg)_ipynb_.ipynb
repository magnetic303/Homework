```python
import gradio as gr
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris
import numpy as np
```

```python
# Logistic Regression
X, y = load_iris(return_X_y=True)
log_reg = LogisticRegression(random_state=0).fit(X, y)

def log_reg_predict(sepal_length, sepal_width, petal_length, petal_width):
  x = np.array([sepal_length, sepal_width, petal_length, petal_width])
  return log_reg.predict(np.array([x]))
```

```python
log_reg_demo = gr.Interface(fn=log_reg_predict, inputs=["number", "number", "number", "number"], outputs="number")
log_reg_demo.launch()
```
