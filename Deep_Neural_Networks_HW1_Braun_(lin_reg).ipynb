```python
import gradio as gr
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
# Linear Regression
X = np.array([[2104, 5], [1416, 3], [852, 2]]) # Size (sqft) and Number of Bedrooms
y = np.array([460, 232, 178]) # Price (1000s dollars)
reg = LinearRegression().fit(X, y)

def reg_predict(sqft, rooms):
  x = np.array([sqft, rooms])
  return reg.predict(np.array([x]))
```

```python
lin_reg_demo = gr.Interface(fn=reg_predict, inputs=["number", "number"], outputs="number")
lin_reg_demo.launch()
```
